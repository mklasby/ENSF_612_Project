```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('runs.csv', header=0)
df
```

```python
model_name = "RandomForestClassifier"
rf = df[df.mlModelClass.str.contains(model_name)]
```

```python
rf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[18,12])
sns.barplot(x="maxDepth", y="f1", hue = "numTrees", data=rf, ax=ax)
fig.show()
plt.savefig("Tuning_RandomForestBar")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[18,12])
sns.lineplot(x="maxDepth", y="f1", hue = "numTrees", data=rf, marker="o", ax=ax)
fig.show()
plt.savefig("Tuning_RandomForestLine")
```
